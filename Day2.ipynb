# Whether a year is a leap year

```python
year=int(input("Enter the year"))
if year%4==0:
    print("The year is leap year")
else:
    print("The year is a normal year")
```

# Palindrome Number

```python
number=int(input("Enter the number"))
number2=number
final_number=0
while(number!=0):
    add_number=number%10
    number=number//10
    final_number=final_number*10+add_number
if (final_number==number2):
    print("The number is pallindrome")
else:
    print("No")
```

# Withdrawing Exercise

```python
balance=5000
count=0
while True:
    pin=1234
    
    user_pin=int(input("Enter your pin"))
    
    if  user_pin ==pin:
        print("Your balance is : ", balance)
        amount=int(input("Enter the amount to withdraw"))
        if balance >= amount:
            print("Amount is withdrawn : ", amount)
            balance-=amount
            print("Current balance : ", balance)
        else:
            print("Invalid input")
            break
    else:
        print("Wrong pin entered")
        break
```
